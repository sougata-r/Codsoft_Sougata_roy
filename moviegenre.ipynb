```python
import pandas as pd
import numpy as np
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv(
    "movietrain_data.txt",
    sep=':::',
    header=None,
    names=['ID', 'TITLE', 'GENRE', 'PLOT']
)
train_df.head()
```

```python
train_df.shape
```

```python
train_df['TITLE'] = train_df['TITLE'].str.strip()
train_df['GENRE'] = train_df['GENRE'].str.strip()
train_df['PLOT'] = train_df['PLOT'].str.strip()
```

```python
top_genres = train_df['GENRE'].value_counts().head(10).index.tolist()
df_filtered_train = train_df[train_df['GENRE'].isin(top_genres)].copy()

X = df_filtered_train['PLOT']
y = df_filtered_train['GENRE']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python

from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
y_train_encoded = lb.fit_transform(y_train)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='liblinear', multi_class='ovr', random_state=42, max_iter=1000)
lr.fit(X_train_tfidf, y_train)
```

```python
y_pred=lr.predict(X_test_tfidf)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import LinearSVC
svc_model = LinearSVC(random_state=42, max_iter=1000)
svc_model.fit(X_train_tfidf, y_train)

y_pred_svc = svc_model.predict(X_test_tfidf)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,y_pred_svc))
print(classification_report(y_test,y_pred_svc))
```
